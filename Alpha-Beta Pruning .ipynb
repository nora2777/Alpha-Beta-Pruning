## 3.2. Alpha-Beta Pruning Algorithm
<font size="4">
Alpha-beta pruning algorithm is an improvement of the minimax algorithm, they both build a
tree that starts from a given state to determine the best move for the player using them. They
use MAX for the player himself as he is trying to maximize his performance and MIN for the
opponent as he is trying to minimize the player performance. Each level of the tree represents
different player moves where even levels represent MAX and odd levels represent MIN.<br>
To build the tree, at the beginning it will start by generating a new state for each legal move for
MAX player, then for each state of them, it will generate a new state for each legal move for
MIN player, and repeat these steps until it reaches the terminal states where the game is ended.<br>
Then evaluation function is calculated for each terminal state and helps the player to determine
which action to take. After that, at MIN level, it will choose from the next level the action that
leads to minimizing the MAX performance which has the lowest evaluation function. At MAX level,
it will choose from the next level the action that leads to maximizing its performance
which has the highest evaluation function. Therefore, we will get the final decision in the root.<br>
However, space and time complexities will be huge for the long tree as in the minimax
algorithm, therefore alpha-beta pruning algorithm came to resolve this problem by pruning the
unnecessary branches which will not affect the result.<br>
To prune the branches, MAX will use α, which initially equals , as the maximum value of the
actions from the next level which will be considered as the lower bound, so it will not take a
value less than α but it will be increased each time it encounters greater value. And MIN will use
β, which initially equals , as the minimum value of the actions from the next level which will be
considered as the upper bound, so it will not take a value greater than β but it will be
decreased each time it encounters smaller value. Cutoff branches at MAX node n if α(n) ≥ β(n),
because whatever was the value of α, MIN would not take it since it always looks for a value
smaller than its β, so there is no need to continue searching in that branch. Cutoff branches at
MIN node n if β(n) ≤ α(n), because whatever was the value of β, MAX would not take it since it
always looks for a value greater than its α, so there is no need to continue searching in that
branch.
</font>

2.3. خوارزمية اقتطاع ألفا-بيتا
</font>   
  <br>
<font size="3">
خوارزمية اقتطاع ألفا-بيتا هي تطوير لخوارزمية minimax. كلاهما يقومان ببناء شجرة تبدأ من حالة معطاه لتحديد أفضل حركة للاعب الذي استخدمها. تستخدم الخوارزميتان اسم MAX للاعب نفسه حيث انه يحاول إعلاء أداءه، واسم MIN للمنافس حيث أنه يحاول إخفاض أداء اللاعب الأول. كل مستوى من الشجرة يمثل حركات لاعب مختلف، فالمستويات الزوجية تمثل MAX، والمستويات الفردية تمثل MIN.<br>
لبناء الشجرة، سيتم في البداية إنشاء حالات جديدة لكل حركة متاحة للاعب MAX، ثم لكل واحدة منهن سيتم إنشاء حالة جديدة لكل حركة متاحة للاعب MIN، ويتم تكرار هذه الخطوات حتى يتم الوصول لحالات التوقف حيث أن اللعبة تنتهي هناك.
ثم يتم حساب دالة التقييم لكل حالة توقف والتي تساعد اللاعب لتحديد الحركة التي سيقوم بفعلها. بعد ذلك في مستوى MIN، سيختار من المستوى التالي الحركة التي تؤدي لإخفاض أداء MAX والتي لديها اقل دالة تقييم. وفي مستوى MAX، سيختار من المستوى التالي الحركة التي تؤدي لإعلاء أداءه والتي لديها اعلى دالة تقييم. وبالتالي سوف نحصل على القرار النهائي في أعلى الشجرة.
مع ذلك، تعقيدات المساحة والوقت سوف تكون هائلة للشجرة كاملة كما في خوارزمية minimax، وبالتالي خوارزمية اقتطاع ألفا-بيتا أتت لتحل هذه المشكلة بواسطة اقتطاع الفروع الغير ضرورية التي لن تؤثر على النتائج.<br>
لاقتطاع الفروع، MAX سيستخدم α، والتي تساوي مبدئياً  -∞، كقيمة عظمى للحركات من المستوى التالي والتي سيعتبرها كحد أدنى، فهو لن يأخذ قيمة أقل من α، ولكن سيزيدها في كل مره يواجه قيمة أكبر. أما MIN سيستخدم β، والتي تساوي مبدئياً  +∞، كقيمة صغرى للحركات من المستوى التالي والتي سيعتبرها كحد أعلى، فهو لن يأخذ قيمة أكبر من β، ولكن سينقصها في كل مره يواجه قيمة أصغر. يتم قص الفروع في n والتي هي MAX، إذا كانت α(n) ≥ β(n)، لأنه مهما كانت قيمة α، لن يأخذها MIN حيث أنه دائماً يبحث عن قيم أصغر من β خاصته، لذلك لا حاجة لإكمال البحث في ذلك الفرع. ويتم قص الفروع في n والتي هي MIN، إذا كانت β(n) ≤ α(n)، لأنه مهما كانت قيمة β، لن يأخذها MAX حيث أنه دائماً يبحث عن قيم أكبر من α خاصته، لذلك لا حاجة لإكمال البحث في ذلك الفرع.<br>


</font>

### 3.2.1. Performance of Alpha-Beta Pruning Algorithm
<font size="4">
The worst case of alpha-beta pruning is when there are no cutoffs performed, it becomes as minimax algorithm with O($b^d$), where b is the branch factor, and d is the depth of terminal state. Whereas the best case is when each player’s best move examined first that leads to cutting off the branches of the tree, it then has O($b^{d/2}$).

</font>

### 3.2.2. Implementation of Alpha-Beta Pruning Algorithm
<font size="4">

Alpha-beta pruning algorithm takes the current state as an input to get the best move for the player who is considered as MAX, it is implemented by first calling the MAX-VALUE function and send as parameter the current state and the initial values of α and β.<br>
MAX-VALUE function checks if the state is terminal, it will return the EVALUATION function of the state, otherwise it will initiate v with -∞ and loop through all successors from the MIN level of the current state to get the maximum value of v or the returned value of MIN-VALUE function. Then it will cut off the search if it encounters that v is greater than β, otherwise it takes the maximum value of α or v as α and continue the search. At the end of the search, it will return the final value of v it got.<br>
MIN-VALUE function also does the same as MAX_VALUE function but in reverse concept, since it tries to minimize the performance. It checks if the state is terminal, it will return the EVALUATION function of the state, otherwise, it will initiate v with +∞ and loop through all successors from the MAX level of the current state to get the minimum value of v or the returned value of MAX-VALUE function. Then it will cut off the search if it encounters that v is lower than α, otherwise it takes the minimum value of β or v as β and continue the search. At the end of the search, it will return the final value of v it got.<br>
The mechanism of the search in the alpha-beta pruning algorithm is recursive, where it calls MAX-VALUE function which in turn calls MIN-VALUE function and vice versa.<br>


</font>

2.2.3. تنفيذ خوارزمية اقتطاع ألفا-بيتا
</font>   
  <br>
<font size="3">
خوارزمية اقتطاع ألفا-بيتا تأخذ الحالة الحالية كمدخل لتحصل على أفضل حركة للاعب والذي تعتبره MAX. الخوارزمية منفذه بواسطة استدعاء دالة MAX-VALUE أولاً ثم ترسل لها الحالة الحالية والقيم الابتدائية لكل من α وβكمتغيرات.<br>
دالة MAX-VALUE تتحقق مما إذا كانت الحالة هي حالة توقف، فتعيد قيمة دالة التقييم (دالة EVALUATION) للحالة، عدا ذلك فستنشئ المتغير v بقيمة -∞ وتدور عبر جميع اللاحقات من مستوى MIN للحالة الحالية لتحصل على القيمة الأكبر بين v أو القيمة المُعادة من دالة MIN-VALUE.<br> ثم تقوم بقص البحث إذا واجهت قيمة v أكبر من قيمة β، عدا ذلك ستأخذ القيمة الأكبر بين α أوv كقيمة α وتكمل البحث. في نهاية البحث، ستعيد الدالة القيمة النهائية التي حصلت عليها في المتغير v.<br>
دالة MIN-VALUE تقوم كذلك بنفس العمليات في دالة MAX_VALUE ولكن بمفهوم عكسي، حيث انها تحاول إخفاض الأداء. تتحقق الدالة مما إذا كانت الحالة هي حالة توقف، فتعيد قيمة دالة التقييم (دالة EVALUATION) للحالة، عدا ذلك فستنشئ المتغير v بقيمة +∞ وتدور عبر جميع اللاحقات من مستوى MAX للحالة الحالية لتحصل على القيمة الأصغر بين v أو القيمة المُعادة من دالة MAX-VALUE. ثم تقوم بقص البحث إذا واجهت قيمة v أقل من قيمة α، عدا ذلك ستأخذ القيمة الأصغر بين β أوv كقيمة β وتكمل البحث. في نهاية البحث، ستعيد الدالة القيمة النهائية التي حصلت عليها في المتغير v.<br>
آلية البحث في خوارزمية اقتطاع ألفا-بيتا هي عَوديّة، حيث أنها تستدعي دالة MAX-VALUE والتي بدورها تستدعي دالة MIN-VALUE والعكس صحيح.

```python
import copy
import sys
class Action:
    def __init__(self, state):
        self.state = state
        self.next_pos = None # next suggested position to mark
        self.v = None 
```


<font size="4">
Class Action contains a constructor that takes a state as input and creates an object with the following attributes:<br>
state: is the current state of the board.<br>
next_pos: is the next suggested position to mark, initially, it is None.<br>
v: is the evaluation value in the case of the terminal state, otherwise, it is the value of the state with the marked next_pos.<br>

</font>






<h2 dir="rtl">
<font size="3">
Class Action يحتوي على مُكوِّن والذي يأخذ الحالة كمُدخل وينشئ كائن بالخصائص التالية:<br>
state: الحالة الحالية للوحة اللعب.<br>
next_pos: المكان التالي المقترح ليتم تمييزه، مبدئياً هي None.<br>
v: قيمة دالة التقييم اذا كانت حالة توقف، عدا ذلك فهو قيمة الحالة مع تمييز next_pos.<br>

</font>

```python
def successor(action, position, turn):
    if action.state[position[0]][position[1]] == " ":
        suc = copy.deepcopy(action.state)
        suc[position[0]][position[1]] = turn
        return Action(suc) # if position is empty, it returns an action with new state
    return None # otherwise it returns None
```


<font size="4">
successor method takes an action, position, and turn as inputs. It checks if the position is not empty in the state of the action, it will return None that means no successor can be generated with that position, otherwise, it will copy the original state in suc variable and mark the position with the turn symbol given, and return a new action containing the newly generated successor state.

</font>



<h2 dir="rtl">
<font size="3">
دالة successor تأخذ حركة، مكان، ودور كمدخلات. تتأكد الدالة مما إذا كان المكان ليس فارغاً في حالة الحركة فسوف تُعيد None والتي تعني لا يوجد لاحق يمكن إنشاءه مع هذه المكان. عدا ذلك ستقوم بنسخ الحالة الأصلية في متغير suc وتقوم بتمييز المكان برمز الدور المعطى، وتُعيد حركة جديدة تحتوي على حالة اللاحق المنشأة حديثاً.
</font>
</h2>

```python
def print_board(state):
    for i in state:
        print (*i, sep =' | ')
```


<font size="4">
This method is used to print a given state (board) separating the columns by ‘|’.
</font>

<h2 dir="rtl">
<font size="3">
يتم استخدام هذه الدالة لطباعة حالة (لوحة) معطاه مع فصل الأعمدة بعلامة ’|’.
</font>
</h2>

```python
def evaluation(state, turn): # return None if not terminal, otherwise return the evaluatuion of the state
    for i in state: # checks rows
        if i[0] == i[1] == i[2]:
            if i[0] == " ":
                continue
            elif i[0] == turn:
                return 1
            else: 
                return -1
    for i in range(3): #checks columns
        if state[0][i] == state[1][i] == state[2][i]:
            if state[0][i] == " ":
                continue
            elif state[0][i] == turn:
                return 1
            else: 
                return -1 
            
    # checks diagonals
    if state[0][0] == state[1][1] == state[2][2] or state[0][2] == state[1][1] == state[2][0]:
        if state[1][1] == " ":
            return None
        elif state[1][1] == turn:
            return 1
        else: 
            return -1
    
    # checks spaces
    for i in state:
        if i.__contains__(" "):
            return None
        
    # no more space and no one wins therefore tie between them
    return 0
```


<font size="4">

evaluation method takes a state and the MAX turn symbol, where it will check if the state is terminal. It will check if there is a winner in either rows, columns, or diagonals, it will determine who the winner is, if he is MAX it will return 1 and if he is MIN it will return -1, otherwise, it will check if the game still has empty squares that mean the game is not ended so it will return None, or it will return 0 since the game has ended (no empty squares left) with no winners so it is a tie between them.


</font>






<h2 dir="rtl">
<font size="3">
دالة evaluation تأخذ حالة ورمز الدور MAX، حيث أنها تتأكد مما إذا كانت الحالة هي حالة توقف. تتأكد الدالة مما إذا كان هناك فائز في أحد الصفوف، الأعمدة، أو الأقطار، عند ذلك ستقوم بتحديد من هو الفائز، إذا كان MAX فستعيد 1، وإذا كان MIN فستعيد -1. عدا ذلك ستتأكد مما إذا كانت اللعبة لازالت تحتوي على مربعات فارغة والتي تعني أن اللعبة لم تنتهي بعد، فبذلك ستعيد None، أو ستعيد 0 حيث أن اللعبة انتهت (لم يبقى هناك مربعات فارغة) بدون فائز لذلك حدث تعادل بينهما.
</font>
</h2>

```python
def MAX(action, a, b, turn): # -> Action
    action.v = evaluation(action.state, turn[0])
    if action.v is not None: # check terminal state
        return action  
    action.v =   float('-inf')
    # loop through successors of current state
    for i in range(3):
        for j in range(3):
            position = [i, j]
            suc = successor(action, position, turn[0])
            if suc is None: # position is not empty (not possible successor)
                continue
            a1 = MIN(suc, a, b, turn)
            #max_v = max(action.v, a1.v)
            if a1.v > action.v: # successor has the maximum v 
                action.next_pos = position # take the position as next_pos
                action.v = a1.v # change v of action to be the v of the successor
            if action.v >= b : # cut off rule
                return action
            a = max(a, action.v)   
    return action
```


<font size="4">
MAX method takes an action, a (α), b (β), and turn list. First, it will check if the action’s state is terminal by invoking the evaluation method sending it the turn at 0 which contains the MAX’s symbol, if it is terminal the MAX method will return the action, otherwise it will initialize the action’s v by –infinity, then it will iterate over all successors of the current state, it will initiate a1 by the returned value from MIN method which takes each successor as an input, then, it will update action’s v if a1’s v (successor’s v) is greater than it and it will set the next_pos with the position at which the successor marked MAX’s symbol. After that, it will cut off the branch and return the current action if the action’s v is greater than or equals to b, otherwise, it will update the value of a with the maximum value between a and the action’s v. At the end of the iterations, if no cutoffs occur, the final action will be returned.


</font>


<h2 dir="rtl">
<font size="3">
دالة MAX تأخد حركة، a (α)، b (β)، وقائمة الأدوار. في البداية ستتحقق مما إذا كانت حالة الحركة هي حالة توقف باستدعاء دالة evaluation وإرسال الدور عند 0 من قائمة الأدوار والتي تحتوي على رمز MAX. إذا كانت حالة توقف، دالة MAX ستعيد الحركة، عدا ذلك ستعطي v التابع للحركة قيمة –مالانهاية، ثم ستدور على جميع اللاحقات للحالة الحالية، وتنشئ a1 بالقيمة المعادة من دالة MIN والتي ستأخذ كل لاحق كمدخل، ثم ستحدث قيمة v التابعة للحركة إذا كانت قيمة v التابعة ل a1(v التابعة للاحق) أكبر منها وستقوم بجعل قيمة next_pos بقيمة المكان الذي قام بتمييزه اللاحق برمز MAX. بعد ذلك ستقوم بقص الفرع وإعادة الحركة الحالية إذا كانت قيمة v التابعة للحركة أكبر من أو تساوي b. عدا ذلك ستقوم بتحديث قيمة a بالقيمة الكبرى بين a وv التابعة للحركة. في نهاية الدورات حيث لم يحدث أي قص سيتم إعادة الحركة النهائية.
</font>
</h2>

```python
def MIN(action, a, b, turn): # -> Action
    action.v = evaluation(action.state, turn[0])
    if action.v is not None: # check terminal state
        return action   
    action.v = float('inf')
    # loop through successors of current state
    for i in range(3):
        for j in range(3):
            position = [i, j]
            suc = successor(action, position, turn[1])
            if suc is None: # position is not empty
                continue
            a1 = MAX(suc, a, b, turn)
            #min_v = min(action.v, a1.v)
            if a1.v < action.v: # successor has the minimum v 
                action.next_pos = position # take the position as next_pos
                action.v = a1.v # change v of action to be the v of the successor
            if action.v <= a : # cut off rule
                return action
            b = min(b, action.v)                
    return action
```


<font size="4">
MIN method also does the same as MAX method but in reverse concept, since it tries to minimize the performance.


</font>






<h2 dir="rtl">
<font size="3">
دالة MIN تقوم كذلك بنفس العمليات في دالة MAX ولكن بمفهوم عكسي، حيث انها تحاول إخفاض الأداء.
</font>
</h2>

```python

def alpha_beta(initial_state, initial_turn): # -> suggested Action 
    if initial_turn == "X": # set list where turn[0] represents MAX, turn[1] represents MIN
        turn = ["X", "O"]
    else:
        turn = ["O", "X"]
    action = MAX(Action(initial_state), float('-inf'), float('inf'), turn)
    if action.next_pos is None: # initial state is terminal state so exit game 
        exit_game(action, turn)
    action.state[action.next_pos[0]][action.next_pos[1]] = initial_turn  # mark the board with suggested move
    return action
    
```


<font size="4">
alpha_beta method takes as parameters the initial_state (current state of the board) and the initial_turn (MAX player symbol). It will order the turn list by making MAX’s symbol at 0 and MIN’s symbol at 1. After that, it will invoke the MAX method and send a new action containing the initial_state, and the initial values of α and β, and the created turn list. Then, it will check if the returned action’s next_pos is not updated with the suggested position to mark that means the initial_state is a terminal, so it will exit the game, otherwise, it will mark and return the state with the new move as a suggested action.
</font>



<h2 dir="rtl">
<font size="3">
دالة alpha_beta تأخذ كمتغيرات الحالة الابتدائية (الحالة الحالية للوحة اللعب) والدور الابتدائي (رمز اللاعب MAX). سترتب قائمة الأدوار بجعل رمز MAX في 0 ورمز MIN في 1.<br> بعد ذلك ستقوم باستدعاء دالة MAX وترسل حركة جديدة تحتوي على الحالة الابتدائية، والقيم الابتدائية ل α وβ، وقائمة الأدوار المنشأة. ثم ستتأكد مما إذا كانت قيمة next_pos للحركة المعادة لم تتحدث بالمكان المقترح للتمييز والذي يعني أن الحالة الابتدائية هي حالة توقف، بذلك ستخرج من اللعبة، عدا ذلك ستقوم بالتمييز وإرجاع الحالة بالحركة الجديدة كحركة مقترحة.

</font>

```python
def exit_game(action, turn): # print the result of the game, and end the game
    if action.v == 0:
        print("Tie")
    elif action.v == 1:
        print(turn[0] + " wins, " + turn[1] + " loses")
    else:
        print(turn[1] + " wins, "+ turn[0] + " loses")
    sys.exit("Game Over")
    
```


<font size="4">
exit_game method will be invoked when the game is ended to determine the winner and the loser players in the game or if there is a tie between them and exit the system.
